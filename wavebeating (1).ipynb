Please write a program to demonstrate wave beating.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up parameters for the waves
t = np.linspace(0, 200, 1000)
frequency1 = 50
frequency2 = 60
amplitude = 1

# Create the two waves
wave1 = amplitude * np.sin(2 * np.pi * frequency1 * t)
wave2 = amplitude * np.sin(2 * np.pi * frequency2 * t)

# Create the resulting wave through superposition
resultant_wave = wave1 + wave2

# Plot the individual waves and the resultant wave
plt.figure(figsize=(10, 6))
plt.plot(t, wave1, label='Wave 1', linestyle='--')
plt.plot(t, wave2, label='Wave 2', linestyle='--')
plt.plot(t, resultant_wave, label='Resultant Wave', linestyle='-', linewidth=2)

plt.xticks(np.arange(0, 201, step=10))
plt.title('Wave Beating Demonstration')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()

```
